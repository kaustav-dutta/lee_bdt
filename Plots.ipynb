```python
d.DRAW_SYS = True
d.DRAW_LEE = False
mode = "bdt"

if mode != "selection":
    p_15 = ROOT.TPaveText(0.74, 0.055, 0.89, 0.1035, "NDC")
    p_white = ROOT.TPaveText(0.78, 0.057, 0.85, 0.091, "NDC")
    p_white2 = ROOT.TPaveText(0.875, 0.055, 0.925, 0.091, "NDC")
    p_3 = ROOT.TPaveText(0.875, 0.057, 0.925, 0.1035, "NDC")
else:
    p_15 = ROOT.TPaveText(0.74, 0.05, 0.79, 0.0835, "NDC")
    p_white = ROOT.TPaveText(0.78, 0.048, 0.83, 0.081, "NDC")
    p_white2 = ROOT.TPaveText(0.875, 0.055, 0.925, 0.081, "NDC")
    p_3 = ROOT.TPaveText(0.875, 0.05, 0.925, 0.0835, "NDC")

p_15.AddText("1.5")
p_15.SetFillStyle(0)
p_15.SetShadowColor(0)
p_15.SetBorderSize(0)
p_15.SetTextFont(42)

p_white.AddText("  ")
p_white.SetFillColor(ROOT.kWhite)
p_white.SetShadowColor(0)
p_white.SetBorderSize(0)
p_white.SetTextFont(42)
p_white.SetTextSize(14)

p_white2.AddText("  ")
p_white2.SetFillColor(ROOT.kWhite)
p_white2.SetShadowColor(0)
p_white2.SetBorderSize(0)
p_white2.SetTextFont(42)
p_white2.SetTextSize(14)

p_3.AddText("3")
# p_3.SetFillColor(ROOT.kWhite)
p_3.SetFillStyle(0)
p_3.SetShadowColor(0)
p_3.SetBorderSize(0)
p_3.SetTextFont(42)
```

```python
c_true, legend = d.plot_variable("reco_energy", mode)
ROOT.gStyle.SetHistTopMargin(0)
if "energy" in c_true.GetName() and not "shower" in c_true.GetName():
    ROOT.gStyle.SetHistTopMargin(0.1)
    p_white.Draw()
    p_white2.Draw()
    p_15.Draw()
    p_3.Draw()
    legend.Draw()
c_true.Draw()
legend.Draw()
c_true.SaveAs("plots/h_%s.pdf" % c_true.GetName()[1:])
```
