```python
d.DRAW_SYS = True
d.DRAW_LEE = True
mode = "cuts"

if mode != "selection":
    p_15 = ROOT.TPaveText(0.69, 0.05806452, 0.7420344, 0.1032258, "NDC")
    p_white = ROOT.TPaveText(0.705, 0.057, 0.925, 0.091, "NDC")
    p_3 = ROOT.TPaveText(0.875, 0.057, 0.925, 0.1035, "NDC")
else:
    p_15 = ROOT.TPaveText(0.685, 0.05, 0.735, 0.0835, "NDC")
    p_white = ROOT.TPaveText(0.68, 0.048, 0.925, 0.081, "NDC")
    p_3 = ROOT.TPaveText(0.875, 0.05, 0.925, 0.0835, "NDC")

p_15.AddText("1.9")
p_15.SetFillStyle(0)
p_15.SetShadowColor(0)
p_15.SetBorderSize(0)
p_15.SetTextFont(42)

p_white.AddText("  ")
p_white.SetFillColor(ROOT.kWhite)
p_white.SetShadowColor(0)
p_white.SetBorderSize(0)
p_white.SetTextFont(42)
p_white.SetTextSize(14)

p_3.AddText("3")
p_3.SetFillColor(ROOT.kWhite)
p_3.SetFillStyle(0)
p_3.SetShadowColor(0)
p_3.SetBorderSize(0)
p_3.SetTextFont(42)
```

```python
c_true, legend = d.plot_variable("pt_ratio", mode)
ROOT.gStyle.SetHistTopMargin(0)
if "energy" in c_true.GetName() and not "shower" in c_true.GetName():
    ROOT.gStyle.SetHistTopMargin(0.1)
    p_white.Draw()
    p_15.Draw()
    p_3.Draw()
    legend.Draw()
c_true.Draw()
legend.Draw()
c_true.Update()
c_true.SaveAs("plots/h_%s_2.pdf" % c_true.GetName()[1:])
c_true.SaveAs("plots/h_%s_2.C" % c_true.GetName()[1:])
```

```python
c_pdg, l_pdg = d.plot_pdg("shower_energy", mode)
c_pdg.Draw()
ROOT.gStyle.SetHistTopMargin(0.1)
c_pdg.SaveAs("plots/h%s_pdg.pdf" % c_pdg.GetName()[1:])
```
